```python
import numpy
import matplotlib.pyplot as plt
```

```python
from warnings import filterwarnings
filterwarnings("ignore")
```

```python
x = ["sales","profit"]
y = [2000,10000]
plt.figure(figsize=(16,4))
plt.bar(x,y,color=["g",'r'],width=0.6)
plt.xlabel("Factor", fontdict={"fontsize":30,"fontfamily":"Times New Roman"})
plt.ylabel("Result")
plt.title("Sales VS Profit" ,size =50, color ="purple")
plt.yticks(size=14)
plt.xticks(size=14)
plt.show()
```

```python
x = ["sales","profit"]
y = [2000,10000]
plt.figure(figsize=(16,4))
plt.barh(x,y,color=["g",'r'])
plt.xlabel("Factor", fontdict={"fontsize":30,"fontfamily":"Times New Roman"})
plt.ylabel("Result")
plt.title("Sales VS Profit" ,size =50, color ="purple")
plt.yticks(size=14)
plt.xticks(size=14)
plt.show()
```
