```python
import numpy
import matplotlib.pyplot as plt

x = ["sales","profit"]
y = [2000,10000]
plt.figure(figsize=(16,4))
plt.plot(x,y,marker ="x", markersize =18)
plt.xlabel("Factor", fontdict={"fontsize":30,"fontfamily":"Times New Roman"})
plt.ylabel("Result")
plt.title("Sales VS Profit" ,size =50, color ="purple")
plt.yticks(size=14)
plt.xticks(size=14)
plt.show()
```

```python
import numpy
import matplotlib.pyplot as plt

x = ["sales","profit"]
y = [2000,10000]
y1 =[4000,8000]
plt.figure(figsize=(16,4))
plt.plot(x,y,marker ="x", markersize =18)
plt.plot(x,y1,marker ="o", markersize =18)
plt.legend([2016,2017])
plt.xlabel("Factor", fontdict={"fontsize":30,"fontfamily":"Times New Roman"})
plt.ylabel("Result")
plt.title("Sales VS Profit" ,size =50, color ="purple")
plt.yticks(size=14)
plt.xticks(size=14)
plt.show()
```
