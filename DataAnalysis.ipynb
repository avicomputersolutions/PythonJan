```python
from csv import reader
```

```python
file = open("sales_data.csv")
readfile = reader(file)

dataList = list(readfile)
dataList
```

```python
# data extraction
columns = dataList[0]
data = dataList[1:]
print(columns)
data
```

```python
saleDict = { columns[0] : [] ,
            columns[1] : [], 
            columns[2] : [], 
            columns[3] : [],
            columns[4] : [],
            columns[5] : []
            } 

saleDict
```

```python
for row in  data :
    saleDict["date"].append(row[0])
    saleDict["product"].append(row[1])
    saleDict["category"].append(row[2])
    saleDict["price"].append(row[3])
    saleDict["quantity"].append(row[4])
    saleDict["revenue"].append(row[5])
```

```python
def unique(data):
    u = set()
    for row in data:
        u.add(row)
    
    return u
```

```python
def isnull(data):
    
    for i in data.keys():
        count = 0
        for j in data[i]:
            if j == '':
                count +=1
                
        print(i , count)

```
