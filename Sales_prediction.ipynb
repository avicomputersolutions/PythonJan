# Data Analysis

What was the total revenue generated by the company over the course of the year?

```python
total =0
for i in data:
    total += i[-1]

print(total)
```

Which product had the highest revenue? How much revenue did it generate?

```python
max=0
for i in data:
    if i[-1]>max:
        max = i[-1]

product = ""
for i in data:
    if i[-1] == max:
        product = i[1]

total_revenue_of_product = 0
for i in data:
    if i[1] == product:
        total_revenue_of_product += i[-1]

print(product, total_revenue_of_product)
```

What was the average price of a product sold by the company?

```python
products = set()
for i in data:
     products.add(i[1])
```

```python
products
```

```python
avgpriceVSproduct = dict()
for i in products:
    sum=0
    count =0
    for j in data:
        if j[1] == i:
            sum += j[-3]
            count +=1
    
    avgpriceVSproduct[i] = sum/count
```
