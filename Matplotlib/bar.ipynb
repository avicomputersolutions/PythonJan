```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("../sales_data.csv")
```

```python
data
```

```python
x = data["category"].unique()
y = data.groupby(by="category")["revenue"].sum()
```

```python
barlabel = plt.bar(x,y)
plt.bar_label(barlabel)
plt.title("Category VS revenue")
plt.xlabel("Category")
plt.ylabel("Revenue")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(12,6))

barlabel = plt.bar(x,y)
plt.bar_label(barlabel)
plt.title("Category VS revenue")
plt.xlabel("Category")
plt.ylabel("Revenue")

plt.show()
```

```python
x = data["category"].unique()
y = data["category"].value_counts()
plt.figure(figsize=(12,6))

barlabel = plt.barh(x,y)
plt.bar_label(barlabel)
plt.title("Category VS revenue")
plt.xlabel("Category")
plt.ylabel("Revenue")

plt.show()
```
