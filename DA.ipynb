```python
# check None 
def checkNone(data, index,col):
    for row in data:
        if row[index] is None:
            print(f"{col[index]} is having None")
            break
    else: 
        print(f"{col[index]} is not having None")
```

```python
checkNone(sales_data,0,columns)
checkNone(sales_data,1,columns)
checkNone(sales_data,2,columns)
checkNone(sales_data,3,columns)
checkNone(sales_data,4,columns)
checkNone(sales_data,5,columns)
```

```python
def isNull(data, index,col):
    for row in data:
        if row[index] == '':
            print(f"{col[index]} is having Blank ")
            break
    else: 
        print(f"{col[index]} is not having Blank")
```

```python
isNull(sales_data,0,columns)
isNull(sales_data,1,columns)
isNull(sales_data,2,columns)
isNull(sales_data,3,columns)
isNull(sales_data,4,columns)
isNull(sales_data,5,columns)
```

+ price , quantity revenue has Missing value

## formulas for calculating Missing Value
+ revenue = price * qunatity
+ price = revenue // quantity
+ quantity  = revenue // price


```python
# Data Cleaning
def fillPrice(data, index):
    for row in data:
        if row[index].isnumeric():
            ""
        else:
            row[index] = int(row[-1]) // int(row[-2])


def fillQuantity(data, index):
    for row in data:
        if row[index].isnumeric():
            ""
        else:
            row[index] = int(row[-1]) // int(row[-3])


def fillRevenue(data, index):
    for row in data:
        if row[index].isnumeric():
            ""
        else:
            row[index] = int(row[-2]) * int(row[-3])
```

```python
fillPrice(sales_data, -3)
```

```python
fillQuantity(sales_data, -2)
```

```python
fillRevenue(sales_data,-1)
```

```python
# Change data type:
def changeType(data,index):
    for i in data :
        if type(i[index]) !='int':
            i[index] = int(i[index])
```

```python
changeType(sales_data,-1)
```

```python
changeType(sales_data,-2)
changeType(sales_data,-3)
```
